```python
import psycopg2
CSV_FILEPATH = 'strokes.csv'
conn = psycopg2.connect(host="localhost",database="t81577", user="postgres", password="postgres")
cur = conn.cursor()
with open(CSV_FILEPATH, 'r') as f:

    next(f) # Skip the header row.
    cur.copy_from(f, 'stroke', sep=',')
conn.commit()
cur.close()
conn.close()
```
